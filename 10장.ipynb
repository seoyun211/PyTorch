```python
from sklearn.feature_extraction.text import CountVectorizer
corpus = [
    'This is last chance.',
    'and if you do not have this chance.',
    'you will never get any chance.',
    'will you do get this one?',
    'please, get this chance',
]
vect = CountVectorizer()
vect.fit(corpus)
vect.vocabulary_
```

```python
vect = CountVectorizer(stop_words=["end", "is" "please", "this"]).fit(corpus)
vect.vocabulary_
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

doc = ['I like machine learning', 'I love deep learning', 'I run everyday']
tfidf_vectorizer = TfidfVectorizer(min_df=1)
tfidf_matrix = tfidf_vectorizer.fit_transform(doc)
doc_distance = (tfidf_matrix * tfidf_matrix.T)

print('유사도를 위한', str(doc_distance.get_shape()[0]), 'x', str(doc_distance.get_shape()[1]), '행렬을 만들었습니다.')
print(doc_distance.toarray())
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
import warnings
warnings.filterwarnings(action='ignore')
import gensim
from gensim.models import Word2Vec

sample = open("C:\\Users\\516-08\\chap10\\data\\peter.txt", "r", encoding='UTF8')
s = sample.read()

f = s.replace("\n", " ")
data = []

for i in sent_tokenize(f):
    temp = []
    for j in word_tokenize(i):
        temp.append(j.lower())
    data.append(temp)

data
```

```python
model1 = gensim.models.Word2Vec(data, min_count=1, vector_size =100, window=5, sg=0)

print("Cosine similarity between 'peter' " + "'wendy' - CBOW : ",
      model1.wv.similarity('peter', 'wendy'))
```

```python
print("Cosine similarity between 'peter' " + "'hook' - CBOW : ", model1.wv.similarity('peter', 'hook'))
```

```python
model2 = gensim.models.Word2Vec(data, min_count=1, vector_size=100, window=5, sg=1)

print("Cosine similarity between 'peter' " + "'wendy' - Skip Gram : ", model2.wv.similarity('peter', 'wendy'))
```

```python
print("Cosine similarity between 'peter' " + "'hook' - Skip Gram : ", model2.wv.similarity('peter', 'hook'))
```

```python
from __future__ import print_function
from gensim.models import KeyedVectors
model_kr = KeyedVectors.load_word2vec_format(
    r"C:\Users\516-08\chap10\data\wiki.ko.vec",
    limit=50000
)
```

```python
find_similar_to = '노력'

for similar_word in model_kr.similar_by_word(find_similar_to):
    print("Word: {0}, Similarity: {1:.2f}".format(
        similar_word[0], similar_word[1]
    ))
```

```python
positive_words = ['동물', '육식동물']
negative_words = ['사람']

# 존재하는 단어만 필터링
positive_exist = [w for w in positive_words if w in model_kr.key_to_index]
negative_exist = [w for w in negative_words if w in model_kr.key_to_index]

similarities = model_kr.most_similar(positive=positive_exist, negative=negative_exist, topn=10)
print(similarities)
```
