```python
import csv
from konlpy.tag import Okt
from gensim.models import word2vec

f = open(r'C:\Users\516-08\chap09\data\ratings_train.txt', 'r', encoding='utf-8')
rdr = csv.reader(f, delimiter='\t')
rdw = list(rdr)
f.close()
```

```python
twitter = Okt()

result = []
for line in rdw:
    malist = twitter.pos(line[1], norm=True, stem=True)
    r = []
    for word in malist:
        if not word[1] in ["Josa","Eomi","Punctuation"]:
            r.append(word[0])
    rl = (" ".join(r)).strip()
    result.append(rl)
    print(rl)
```

```python
with open("NaverMovie.nlp", 'w', encoding='utf-8') as fp:
    fp.write("\n".join(result))
```

```python
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler, MinMaxScaler 

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
df = pd.read_csv('C:\\Users\\516-08\\chap09\\data\\diabetes.csv')
X = df[df.columns[:-1]]
y = df['Outcome']

X = X.values
y = torch.tensor(y.values)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
ms = MinMaxScaler()
ss = StandardScaler()

X_train = ss.fit_transform(X_train)
X_test = ss.fit_transform(X_test)
y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)
y_train = ms.fit_transform(y_train)
y_test = ms.fit_transform(y_test)
```

```python
class customdataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y
        self.len = len(self.X)
    def __getitem__(self, index):
        return self.X[index], self.y[index]
    def __len__(self):
        return self.len
```

```python
train_data = customdataset(torch.FloatTensor(X_train), torch.FloatTensor(y_train))
test_data = customdataset(torch.FloatTensor(X_test), torch.FloatTensor(y_test))

train_loader = DataLoader(dataset=train_data, batch_size=32, shuffle=True)
test_loader = DataLoader(dataset=test_data, batch_size=32, shuffle=False)
```

```python
class binaryClassification(nn.Module):
    def __init__(self):
        super(binaryClassification, self).__init__()
        self.layer_1 = nn.Linear(8, 64, bias=True)
        self.layer_2 = nn.Linear(64, 64, bias=True)
        self.layer_out = nn.Linear(64, 1, bias=True)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(p=0.1)
        self.batchnorm1 = nn.BatchNorm1d(64)
        self.batchnorm2 = nn.BatchNorm1d(64)

    def forward(self, inputs):
        x = self.relu(self.layer_1(inputs))
        x = self.batchnorm1(x)
        x = self.relu(self.layer_2(x))
        x = self.batchnorm2(x)
        x = self.dropout(x)
        x = self.layer_out(x)
        return x
```

```python
epochs =20+1
print_epoch = 20
LEARNING_RATE = 1e-2

model = binaryClassification()
model.to(device)
print(model)
BCE = nn.BCEWithLogitsLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=LEARNING_RATE)
```

```python
def accuracy(y_pred, y_test):
    y_pred_tag = torch.round(torch.sigmoid(y_pred))
    correct_results_sum = (y_pred_tag == y_test).sum().float()
    acc = correct_results_sum/y_test.shape[0]
    acc = torch.round(acc * 100)
    return acc
```

```python
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 모델과 손실 함수, 옵티마이저를 device에 맞게 이동
model = model.to(device)
BCE = BCE.to(device) if hasattr(BCE, 'to') else BCE  # BCE가 nn.Module일 경우

for epoch in range(epochs):
    iteration_loss = 0.
    iteration_accuracy = 0.

    model.train()
    for i, data in enumerate(train_loader):
        X, y = data
        X = X.float().to(device)
        y = y.to(device)

        optimizer.zero_grad()
        y_pred = model(X)
        loss = BCE(y_pred, y.reshape(-1,1).float())
        loss.backward()
        optimizer.step()

        iteration_loss += loss.item()
        iteration_accuracy += accuracy(y_pred, y)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

for epoch in range(epochs):
    iteration_loss = 0.
    iteration_accuracy = 0.

    model.train()
    for i, data in enumerate(train_loader):
        X, y = data
        X = X.float().to(device)
        y = y.to(device)

        optimizer.zero_grad()
        y_pred = model(X)
        loss = BCE(y_pred, y.reshape(-1,1).float())
        loss.backward()
        optimizer.step()

        iteration_loss += loss.item()  # .item() 추가
        iteration_accuracy += accuracy(y_pred, y)

    if epoch % print_epoch == 0:
        print('Train: epoch: {0} - loss: {1:.5f}; acc: {2:.3f}'.format(
            epoch, iteration_loss/(i+1), iteration_accuracy/(i+1)))

    iteration_loss = 0.
    iteration_accuracy = 0.
    model.eval()
    with torch.no_grad():  # 테스트 시 gradient 계산 방지
        for i, data in enumerate(test_loader):
            X, y = data
            X = X.float().to(device)
            y = y.to(device)

            y_pred = model(X)
            loss = BCE(y_pred, y.reshape(-1,1).float())

            iteration_loss += loss.item()  # .item() 추가
            iteration_accuracy += accuracy(y_pred, y)

    if epoch % print_epoch == 0:
        print('Test: epoch: {0} - loss: {1:.5f}; acc: {2:.3f}'.format(
            epoch, iteration_loss/(i+1), iteration_accuracy/(i+1)))
```
